```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define the number of iterations and the equal time duration for each iteration
num_iterations = 10000
dt = 0.01  # Small time increment

# Define variables to store particle positions
x = np.zeros(num_iterations)
y = np.zeros(num_iterations)

# Implement Brownian motion
for i in range(1, num_iterations):
    dx = np.random.normal(0, np.sqrt(2 * dt))
    dy = np.random.normal(0, np.sqrt(2 * dt))
    x[i] = x[i - 1] + dx
    y[i] = y[i - 1] + dy
```

```python
# Display the results
plt.plot(x , y)
plt.title('Brownian Motion Model')
plt.xlabel('Position x')
plt.ylabel('Position y')
plt.show()
```
